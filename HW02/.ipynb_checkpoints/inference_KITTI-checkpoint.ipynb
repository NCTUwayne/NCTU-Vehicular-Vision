```python
import os
import time
import cv2
import pickle
import random

import numpy as np
import matplotlib.pyplot as plt
import keras.backend as K
import tensorflow as tf


from PIL import Image
from random import shuffle
from enum import Enum
from keras.models import Model, load_model
from keras.callbacks import EarlyStopping, ModelCheckpoint, TensorBoard, CSVLogger, LearningRateScheduler, TerminateOnNaN
from keras.optimizers import SGD, Adam, RMSprop
from keras.applications.mobilenetv2 import MobileNetV2
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas

from modelDesign import MNv2_depth_multiloss_model
from utils_KITTI import *
```

## Parameters Setting

```python
# 預測深度圖是否為log depth?
log_depth = False

# 欲載入 pretrain weight 的路徑
load_depthWeight_path = '../../models/mnv2_depth_multiloss/berhuLoss_PSMNet_epoch-29_loss-2.4079.h5'

# Dataset路徑
kitti_depth_folder = '../../dataset/KITTI/depth/val/'

# Model Input 解析度設定
INPUT_HEIGHT = 320
INPUT_WIDTH = 1056
```

## Build a new model

```python
# 建立 Model layer 架構
model = MNv2_depth_multiloss_model(inputShape=(INPUT_HEIGHT,INPUT_WIDTH,3), alpha=0.35, 
                                   expansion_factor=6, depth_multiplier=1, 
                                   lock_backend_weights = False)
# 載入Model weight
model.load_weights(load_depthWeight_path)
    
# Model 資訊總覽
model.summary()
```

## Inference on KITTI Dataset

```python
# 載入 validation data 影像的路徑
validationData = getKittiPath(kitti_depth_folder)
```

```python
# 設定預測深度圖的最小值、最大值
depth_min, depth_max = 4,100

# 隨機從 validation list 取一深度圖的 path
depth_path = validationData[random.randint(0,len(validationData)-1)]
print(depth_path)

# 將 depth_path 轉換為對應影像的 image_path
subPath = depth_path.split(os.sep)
image_path = '../../dataset/KITTI/image/'+subPath[6][:10]+'/'+subPath[6]+'/'+subPath[-2]+'/data/'+subPath[-1]

# 讀取影像
image = read_KITTI_image(image_path, shape = (INPUT_HEIGHT, INPUT_WIDTH))

# 建立深度圖的 Variable，將用來承接 model 的預測結果
depth = [np.zeros((20,66)),
         np.zeros((40,132)),
         np.zeros((80,264)),
         np.zeros((160,528))]

# 把每個 resolution 的預測結果，放入 variable: depth 中，並進行一些後處理以便顯示
for i in range(4): 
    
    # 紀錄 Model Infenrece 起始時間
    st = time.time()
    
    # 取得該 resolution 的 predicted depth map
    depth[i] = np.squeeze(model.predict(np.expand_dims(image, axis=0))[i])
    
    # 紀錄 Model Infenrece 結束時間
    et = time.time()
    
    # 顯示 Inference 時間與 FPS
    print('avg time: ', (et-st)/3)
    print('avg fps: ', 1/((et-st)/3))

    
    # 將預測深度圖以 depth_min, depth_max 進行 clipping，
    depth[i][depth[i]>depth_max]=depth_max
    depth[i][depth[i]<depth_min]=depth_min
    
    # 進行log transform，以便顯示
    if not log_depth:
        depth[i] = (np.log2(depth[i])-np.log2(depth_min))*100 / (np.log2(depth_max)-np.log2(depth_min))

    # 將深度圖進行normalization並inverse，以利顯示
    depth[i] *=255./depth_max
    depth[i] = 255-depth[i]

# 讀入影像
image = cv2.resize(image,(1056,320))

# 顯示結果
showResult([image, depth[3], depth[2], depth[1], depth[0]], figsize=(10,15))
```

## Inference on NCTU Videos

```python
def model_predict(image):
    image = cv2.resize(image,(INPUT_WIDTH,INPUT_HEIGHT))
    depth = np.squeeze(model.predict(np.expand_dims(image, axis=0))[3])
    
    depth_min = 1
    depth_max = 80
    depth[depth>depth_max]=depth_max
    depth[depth<depth_min]=depth_min    
    
    if not log_depth:
        depth = (np.log2(depth)-np.log2(depth_min))*depth_max / (np.log2(depth_max)-np.log2(depth_min))
    
    depth = 255*(depth) / (depth.max())
    depth = 255 - depth
    
    return depth
```

```python
# 欲讀取的交大影片 [TODO]
data_keyword = 'V_20180608_163423_vHDR_On'

# 欲預測深度的 frame [TODO]
frame_ID = 250

# 讀取影片
vidcap = cv2.VideoCapture('../../dataset/NCTU/campus/'+data_keyword+'.mp4')

# 取第 frame_ID 個 frame 來預測
for i in range(frame_ID):
    success,image = vidcap.read()

# 預測深度圖
depth = model_predict(image)

# 重新調整影像、深度圖的解析度
depth = cv2.resize(depth,(640,320))
image = cv2.resize(image,(640,320))

# 顯示結果
showResult([image[:,:,::-1], depth], figsize=(15,10))
```

## Calculate RMSE per distance on KITTI Dataset

```python
error_range = 1
total_error_range = 100
rmse_all = [[] for i in range(total_error_range)]


totalTime =0
for dataIter, data in enumerate(validationData[:100]):
    print('\r processing '+str(dataIter), end='')
    
    depth_path = data
    subPath = depth_path.split(os.sep)
    image_path = '../../dataset/KITTI/image/'+subPath[6][:10]+'/'+subPath[6]+'/'+subPath[-2]+'/data/'+subPath[-1]
    
    
    # Load depth
    depth =depth_read_KITTI(depth_path, shape=(375,1242))

    # Load image                
    image = read_KITTI_image(image_path, shape=(INPUT_HEIGHT, INPUT_WIDTH))
    image_expend = np.expand_dims(image, 0)

    # predict
    start = time.time()
    results = model.predict(image_expend,batch_size=1)[0]
    end = time.time()
    totalTime +=end-start
    
    results = cv2.resize(results[0,:,:,:],(1242,375))
    
    # evaluate
    for i in range(total_error_range):
        value = compute_RMSE(results, depth, cap=(i+1)*error_range, error_range=error_range)
        if value!=0:
            rmse_all[i].append(value)

error_per_distance = []
for i in range(total_error_range):
    error_per_distance.append(float(str(sum(rmse_all[i])/(len(rmse_all[i])+0.0000001))))
    #print('%.3f' %float(str(sum(rmse_all[i])/(len(rmse_all[i])+0.0000001))))

print('totalTime: %.2f'  %totalTime)
print('fps: %.2f'        %float(len(validationData)/totalTime))
```
